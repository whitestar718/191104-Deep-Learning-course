### ogird, mgrid

선호도 차이다. mgrid가 더 좋으면 그거 써도 된다.

```python
XX, YY = np.ogrid[10:40:10, 1:4]
```

```python
print(XX)
print(YY)
```

```python
XX, YY = np.mgrid[10:40:10, 1:4] # mesh 그리드의 단축표현이라고 생각해도 된다.
```

```python
print(XX)
print(YY)
```

간혹 책에 j가 등장하기도 한다.

```python
XX, YY = np.mgrid[10:40:10, 1:4:8j] # nj는 몇 개를 만들어 줄 것이냐 이다. j앞에 숫자 바꿔가면서 해봐라.
```

```python
print(XX)
print(YY)
```

이제 아래 식을 이해해 봐라. mgrid와 meshgrid의 차이가. 외울 수밖에 없다.

```python
d1, d2, d3 = np.mgrid[0:10, 0:10, 0:10]
d11, d22, d33 = np.meshgrid(np.arange(10), np.arange(10), np.arange(10))
np.array_equal(d1, d11)
```

```python
d11 # m이 하나니까 동일하다는 개념으로 외우자.
```

```python
d1, d2 = np.mgrid[0:10, 0:10]
d11, d22 = np.meshgrid( np.arange(10), np.arange(10))
np.array_equal(d1, d11)
```

```python
d1 # mgrid 결과 -> 가장 차원이 높은 애쪽에서 만들어준다. 2차원이기 때문에 0, 1, 2, 3,,,, 은 열방향으로 나있다.
# 1차원이면 행방향으로 0, 1, 2가 진행될 것이다.
```

### einsum 원만들기

회전변환행렬을 사용한다.

```python
th = np.linspace(0, 2*np.pi, 1024)
mats = np.array([[np.cos(th), np.sin(th)], [-np.sin(th), np.cos(th)]])
vecs = np.array([1, 0])
vecs.repeat(1024)
x, y = np.einsum('...ij,...i', mats, vecs)
```

```python
y.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
plt.plot(x, y)
```

```python
aa = np.array([[1, 2, 3], [4, 5, 6]])
bb = np.array([[3, 4, 5], [7, 8, 9]])
```

```python
print(aa)
print(bb)
```

```python
np.einsum('...i', aa)
```

```python
np.einsum('...ij', aa)
```

```python
np.einsum('...i, ...j', aa, bb)
```

```python
print(aa)
print(bb)
```

```python
np.einsum('...ij, ...j', aa, bb) # ij뜻은 행을 다 돌아라, 열을 다 돌아라라는 뜻이다.
```

```python
print(y)
```
