# YOLO

요새 디텍션하면 다 욜로다. 근데 학습하기가 좀 귀찮다. 이 애도 진짜 어노테이션이 귀찮다. 라벨링 하는거 말이다. 계보를 좀 보자면

 - R-CNN -> Fast R-CNN -> Faster R-CNN, YOLO -> SSD로 간다.
 
그러니 공부를 하자면, 욜로와 SSD로 공부하는게 좋겠군. 욜로가 더 빠르게 감지하지만, SSD의 mAP가 더 좋다. 디텍션의 핵심은 2개다. 로컬라이제이션과 클래시피케이션인데, 로컬라이제이션을 잘해야 한다. 왜냐면 클래시피케이션은 이미 게임 끝났기 때문. 그럼 욜로의 로컬라이제이션은 어떤식으로 되나? 약간 다르다.

일단 클래시피케이션으로 구글넷을 했다. 근데 욜로 버전이 올라가면 다크넷이라는 것을 쓴다. 그 다음에 컨볼루션, 렐루 쓴다. 풀리 커넥티드 쓰는 것도 똑같다. (풀리커넥티드는 플래튼, 글로벌에버리지 풀링) 근데, 로컬라이제이션을 어떻게 하느냐가 핵심이다. 이건 Fast R-CNN하고 좀 비슷하다. Proposal reason network에서 9개의 예비 후보를 정하게 되는데, 이것과 상당히 비슷하다. 로컬라이제이션 속도를 빠르게 하기 위해서, 어떤 방법을 쓸까...

좀 더 앞선 클래시피케이션을 설명하자면, 파인튜닝이긴 한데 뒤에 컨볼루션 레이어를 아주 많이 추가했다. 더 필요했기 때문. 더 갖다 붙이면 성능은 더 좋아지지만 학습은 잘 안될텐데... 다행히도 이놈은 학습이 잘 되었다고 한다. 

### 더 빠른 로컬라이제이션

그리드 셀을 좀 더 느슨하게 나누어서 7 x 7로 탐색을 하는데, 이러면 속도는 빠르지만 정확도는 좀 떨어진다. 특징이 포함될 수 있는 영역들에 5개의 값들을 하나씩 넣는다. (x, y, w, h, c 자세한 설명은 ppt 자료 참조) 이런식으로 하면 6개정도(?) 가능한 영역이 도출된다. 음... 정확히 말하자면 처음 5, 5, 개는 위의 x, y ... 등의 값이고, 나머지 20개는 클래스값이다. (30개) 이런식으로 49개를 돌아버리는데, 그래서 속도가 빠를 수 밖에 없다. 그러므로 쿠다 개수가 49개면? 한방에 해결되는 것이다. 시간 걸릴 요건이 없다. 정리하면 컨볼루션은 구글넷이 하고, 로컬라이제이션에서 7 x 7로 줄여서 서칭한 것이다. (RCN은 9개를 계산했다. 앵커라고 해서 9개가 돌아다니는것 다시 기억)

이제 opencv욜로 파일로 가서 공부해라. 폴더 있음
